```python
from happytransformer import HappyGeneration
from happytransformer import GENTrainArgs
```

```python
happy_gen = HappyGeneration("GPT-NEO", "EleutaherAI/gpt-neo-125M")
```

## Обучение на шутках 1

```python
train_0 = "/content/drive/Shareddrives/ML/AIIJC/1.1 текст для обучения/train_0_jokes.txt"

train_args = GENTrainArgs(learning_rate = 3e-6, num_train_epochs = 1)
happy_gen.train(train_0, args = train_args)
```

```python
happy_gen.save("/content/drive/Shareddrives/ML/AIIJC/2.1 GPT-Neo/jokes_0")
```
